## Pandas Plot

```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'],columns=iris['feature_names'])
y = pd.DataFrame(iris['target'],columns=['target_names'])
data = pd.concat([x,y],axis=1)
data.head()
```

```python
data.plot.scatter(x="sepal length (cm)",y='sepal width (cm)',c='target_names')
```

```python
color = {
    0:'r',
    1:'g',
    2:'b'
}
```

```python
data['color'] = data['target_names'].map(color)
data.head()
```

```python
data.plot.scatter(x="sepal length (cm)",y='sepal width (cm)',c='color')
```

```python
data[data['target_names']==0].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='r',label='setosa')
data[data['target_names']==1].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='g',label='versicolor')
data[data['target_names']==2].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='b',label='virginica')
```

```python
ax = data[data['target_names']==0].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='r',label='setosa')
ax = data[data['target_names']==1].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='g',label='versicolor',ax = ax)
data[data['target_names']==2].plot.scatter(x="sepal length (cm)",y='sepal width (cm)',color='b',label='virginica',ax = ax)
```

## Plot Style

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2 # 0 to 15 point radii

plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

```python
plt.style.use('ggplot')
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2 # 0 to 15 point radii

plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

```python
plt.style.use('classic')
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2 # 0 to 15 point radii

plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```
